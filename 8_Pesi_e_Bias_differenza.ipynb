PESI (WEIGHTS) E BIAS - I PARAMETRI APPRENDIBILI DELLA RETE

IN UN LAYER LINEARE: y = x * W^T + b
dove:
- x: input (features)
- W: pesi (weights)
- b: bias
- y: output

1. PESI (WEIGHTS) - W
- Sono i coefficienti che moltiplicano l'input
- Determinano "quanto" ogni feature di input contribuisce all'output
- In una matrice di forma (out_features, in_features)
- Ogni neurone ha un vettore di pesi (uno per ogni input)

Esempio concreto:
layer = nn.Linear(784, 512)
layer.weight.shape: (512, 784)
- 512 neuroni nel layer
- Ogni neurone ha 784 pesi (uno per ogni pixel in input)
- peso[neurone12, pixel345] = quanto il pixel345 influenza il neurone12

2. BIAS - b
- È un termine costante aggiunto all'output
- Permette al neurone di attivarsi anche quando l'input è zero
- Vettore di forma (out_features,)
- Ogni neurone ha il suo bias

Esempio:
layer.bias.shape  # (512,) - un bias per ogni neurone

DIFFERENZA CHIAVE:

|                   | PESI (WEIGHTS)                      | BIAS                     |
|-------------------|-------------------------------------|--------------------------|
| Ruolo             | Amplificano/attenuano l'input       | Spostano l'attivazione   |
| Formula           | Moltiplicano l'input (x * W)        | Si sommano (+ b)         |
| Forma             | (out_features, in_features)         | (out_features,)          |
| Numero parametri  | out_features × in_features          | out_features             |
| Inizializzazione  | Di solito casuale (piccoli valori)  | Di solito zero           |
| Effetto           | Cambia la "pendenza" della risposta | Cambia la "soglia"       |

ANALOGIA: UN TERMOSTATO
- Pesi = quanto giri la manopola (sensibilità alla temperatura)
- Bias = la temperatura di base che vuoi mantenere

ANALOGIA: UNA FUNZIONE LINEARE y = mx + q
- m (pendenza) = pesi
- q (intercetta) = bias

TOTALE MODELLO: circa 669.706 parametri apprendibili

PERCHÉ SONO IMPORTANTI:
- Durante l'addestramento, il modello AGGIORNA pesi e bias
- L'obiettivo è trovare i valori che minimizzano l'errore
- La backpropagation calcola come modificare ogni peso e bias
- L'optimizer applica queste modifiche

IN SINTESI: I pesi determinano "cosa guardare" nell'input, i bias determinano "quando attivarsi". Insieme, definiscono completamente il comportamento di ogni neurone.
